Так же как и в колонки `critic_score`  в колонке `user_score` там где рейтинг еще не указан  сборы в основном по этим играм равняются 0.

```python
# отфильттруем таблицу от строк с пустыми значениями в колонках critic_score и user_score и после выведим таблицу корреляций
test = df_games[~(df_games['critic_score'].isna())& ~(df_games['user_score'].isna()) & (df_games['user_score'] != 'tbd')]
test['user_score'] = test['user_score'].astype('float')
test.corr()
```

```python
df_games[(df_games['critic_score'].isna())& ~(df_games['user_score'].isna()) & (df_games['user_score'] != 'tbd')].shape
```

Обе колонки имеют слабую связь с другими колонками а между собой имеют заметную положительную корреляцию. Имея только эту информацие не возможно корректно запольнить пропуски в колонках `critic_score` и `user_score` так как сильной зависимости обнаружено с другими значениями не было а заполнение средним значением или нулевым может привести к значительному изменению анализа в дальнейшем. Поэтому оставим пропуски в этих колонках без изменения. А для перевода колонки `user_score` в формат float из object заменем все значения `tbd` на nan

```python
def nan_user_score(value):
    if value == 'tbd':
        return  np.nan
    else:
        return value
df_games['user_score'] = df_games['user_score'].apply(nan_user_score)
```

```python
df_games['user_score'] = df_games['user_score'].astype('float64')
```

```python
# Проанализируем колонку rating, посмотрим какие уникальные значения в ней встречаются
df_games['rating'].unique()
```

```python
# создадим словарь и посмотрим зависит рейтинг от организации ESRB и возрастная группа от жанра игры.
genre_na= list(df_games[df_games['rating'].isna()]['genre'].unique())
dic_genre = {}
for i in genre_na:
    dic_genre[i] = list(df_games[df_games['genre'] == i]['rating'].unique())
    
dic_genre
```

### Посчитаем суммарные продажи для каждой игры и создадим отдельную колонку с эти значением

```python
def sum_benefit(row):
    benefit = row['na_sales'] + row['eu_sales'] + row['jp_sales'] + row['other_sales']
    return benefit

df_games['sum_benefit'] = df_games.apply(sum_benefit, axis=1)
```

## Проверка гипотез

### Средние пользовательские рейтинги  платформ Xbox One и PC одинаковые

```python
df_platform = df_games_actual[(df_games_actual['platform'] == 'XOne') | (df_games_actual['platform'] == 'PC')].pivot_table(index='platform', values='user_score', aggfunc=['mean', 'var', 'std'])
df_platform
```

```python
df_xone_uscore = list(df_games_actual[(df_games_actual['platform'] == 'XOne') & ~(df_games_actual['user_score'].isna())]['user_score'])
df_pc_uscore = list(df_games_actual[(df_games_actual['platform'] == 'PC') & ~(df_games_actual['user_score'].isna())]['user_score'])
```

Так как нам нужно сравнить средний пользовательский рейтинг в зависимости от  двух платформ то сформулируем нулевую гипотезу так:

H₀ - средний пользовательский рейтинг платформ Xbox One и PC равны

Тогда альтернативная гипотеза будет звучать так:

H₁ - средний пользовательский рейтинг платформ Xbox One и PC отличается между собой

Уровень значимости (alpha) примем раным 0.05
Так как дисперсии по  платформам сильно отличаются между собой примем значение equal_var = False при расчете p-value

```python
# расчет значения p-value между двумя выборками
alpha = .05
result_xone_pc = st.ttest_ind(df_xone_uscore, df_pc_uscore, equal_var = False)
print('p - значение:', result_xone_pc.pvalue)
if result_xone_pc.pvalue < alpha:
    print('Отвергаем нулевую гипотизу')
else: 
    print('Не получилось отвергнуть нулевую гипотизу')
```

### Средние пользовательские рейтинги  жанров Action и Sports одинаковые

```python
df_genre = df_games_actual[(df_games_actual['genre'] == 'Sports') | (df_games_actual['genre'] == 'Action')].pivot_table(index='genre', values='user_score', aggfunc=['mean', 'var', 'std'])
df_genre
```

```python
df_sports_uscore = list(df_games_actual[(df_games_actual['genre'] == 'Sports') & ~(df_games_actual['user_score'].isna())]['user_score'])
df_action_uscore = list(df_games_actual[(df_games_actual['genre'] == 'Action') & ~(df_games_actual['user_score'].isna())]['user_score'])
```

```python
len(df_sports_uscore)
```

Так как нам нужно сравнить средний пользовательский рейтинг в зависимости от двух жанров то сформулируем нулевую гипотезу так:

H₀ -  средний пользовательский рейтинг жанров Action и Sports равны

Тогда альтернативная гипотеза будет звучать так:

H₁ - средний пользовательский рейтинг жанров Action и Sports отличаются между собой

Уровень значимости (alpha) примем раным 0.05
Так как дисперсии каждого жанра сильно отличаются между собой примем значение equal_var = False при расчете p-value

```python
# расчет значения p-value между двумя выборками
alpha = .05
result_sports_action = st.ttest_ind(df_sports_uscore, df_action_uscore, equal_var = False)
print('p - значение:', result_sports_action.pvalue)
if  result_sports_action.pvalue < alpha:
    print('Отвергаем нулевую гипотизу')
else: 
    print('Не получилось отвергнуть нулевую гипотизу')
```
