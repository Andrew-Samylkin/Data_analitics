## Предобработка данных

```python
# переведем значения в колонке total_images в Int64 ( нельзя загрузить половину фотографии)
df_estate['total_images'] = df_estate['total_images'].astype('Int64')
```

```python
# переведем значения в колонке last_price в Int64 так как нет дробной части и в колонки только целые значения
df_estate['last_price'] = df_estate['last_price'].round().astype('Int64')
```

```python
# округлим значения столбца total_erea до 2 знаков после запятой ( в диду того что квадрат нынче дорог влияют даже сотые)
df_estate['total_area'] = df_estate['total_area'].round(2)  
```

```python
# приведем значение колонки first_day_exposition к формату даты  так как так удобнее
df_estate['first_day_exposition'] = pd.to_datetime(df_estate['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
df_estate['first_day_exposition'].head(10)
```

```python
# переведем значения в колонке rooms в Int64 ( int64 не поддерживает операции с пропусками)
df_estate['rooms'] = df_estate['rooms'].astype('Int64')
```

```python
# посмотри на состав уникальных значений в колонке locality_name
df_estate['locality_name'].unique()
```

Можно сделать вывод что среди всего перечня имен населенных пунктов очень много неявных дубликатов в основном это связано с тем что у одних и тех же населенных пунктов присутствуют разные 'приставки'  и написание слов в эти приставках (ё вмсто е).

```python
# так как строк без указания locality_name 49 и их удаление не сильно повлияет на весь датафрейм удалим их.
df_estate.dropna(subset=['locality_name'], inplace=True)
```

```python
def upper(title):
    ''' везвращает только слова написанные с болльшой буквы'''
    string = ' '.join(word for word in title.split(' ') if word[0].isupper())
    return string

#создадим столбец name без приставки к названию населенного пункта
df_estate['flt_name'] = df_estate['locality_name'].apply(upper)
```

```python
df_estate['flt_name'] = df_estate['flt_name'].astype('category')
```

```python
df_estate['flt_name'].head(10)
```

```python
# создадим дата фрейм отфильтровав все некорректные значения по параметру ceiling_height
df_test = df_estate.query('abs(ceiling_height) > 0')
```

```python
# посмотрим как значения в колонки ceiling_height  коррелируют со стральными количественными значениями построив heatmap.
sns.set(rc = {'figure.figsize':(20,8)})
corrMatrix = df_test.corr().round(3)
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
def ceiling_median(row):
    ceiling = row['ceiling_height']
    floors_total = row['floors_total']
    if np.isnan(ceiling):
        ceiling = df_estate[df_estate['floors_total'] == floors_total]['ceiling_height'].median()
        return ceiling
    else:
        return ceiling

df_estate['ceiling_height'] = df_estate.apply(ceiling_median, axis=1)
```

```python
# заполняем все пропущенные значения в колонке living_area
def living_area_fill(row):
    total = row['total_area']
    living = row['living_area']
    if np.isnan(living):
        living = total / k_living_area
        return living
    else:
        return living

df_estate['living_area'] = df_estate.apply(living_area_fill, axis=1)
```
